```python
import pandas as  pd
data={
    "StudyHours":[1,2,3,4,5,6,7,8],
    "Marks":[50,55,65,70,75,85,90,95]
}
df=pd.DataFrame(data)
```

```python
X=df[["StudyHours"]]
y=df["Marks"]
# print(X)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,y)
```

```python
prediction=model.predict([[9]])
print(prediction)
```

```python
#To visualize model
import matplotlib.pyplot as plt
plt.scatter(X,y)
plt.plot(X,model.predict(X))
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study vs marks")
plt.show()
```

```python
#My First ML challenge
data1={
    "Experience":[1,2,3,4,5,6],
    "Salary":[30000,35000,40000,45000,50000,55000]
}
df1=pd.DataFrame(data1)
print(df1)
```

```python
X=df1[["Experience"]]
y=df1["Salary"]
```

```python
model1=LinearRegression()
model1.fit(X,y)
prediction1=model1.predict([[7]])
print(prediction1)
```

```python
plt.scatter(X,y)
plt.title("Experience vs Salary")
plt.plot(X,model1.predict(X))
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```
