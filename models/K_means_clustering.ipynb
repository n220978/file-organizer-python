```python
import pandas as pd
data={
    "Age":[25,28,35,40,45],
    "Income":[30000,32000,60000,70000,90000]
}
df=pd.DataFrame(data)
```

```python
X=df[["Age","Income"]]
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=2)
model.fit(X)
```

```python
clusters=model.labels_
print(clusters)
```

```python
df["Cluster"]=model.labels_
print(df)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["Age"],df["Income"],c=df["Cluster"])
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```
